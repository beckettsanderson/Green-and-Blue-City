```python
import pandas as pd
import openpyxl
import altair as alt
import json
from shapely.geometry import shape, Point
```

Source: https://stackoverflow.com/questions/20776205/point-in-polygon-with-geojson-in-python 

Source for Boston GEOJSON file: https://data.boston.gov/dataset/boston-neighborhoods/resource/60b0a93d-ac1a-4caa-9cc0-91c23717d84f

```python
pts = pd.read_csv("Cleaned Data/Trees_clean.csv")

with open("boston.geojson") as boston_geo:
    js = json.load(boston_geo)
    
    for i in range(len(pts)):
        pt = pts.iloc[i]
        pt_obj = Point(pt.Longitude, pt.Latitude)
        for feature in js["features"]:
            polygon = shape(feature['geometry'])
            if polygon.contains(pt_obj):
                print("Found containing polygon: ", feature["properties"]["Name"])
                pts.loc[i,"Region"] = feature["properties"]["Name"]
```
